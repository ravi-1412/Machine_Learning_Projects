```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
print(x.shape)
print(y.shape)
```

## !pip install imblearn = libary to handle imbalance dataset

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
over_sample =  RandomOverSampler(random_state=42)
```

```python
X_train_res, y_train_res = over_sample.fit_resample(x, y)
```

```python
X_train_res.shape,y_train_res.shape
```

## training and testing model with libary called PYCARET WITH 100000 SAMPLES

```python
df  = pd.concat((X_train_res,y_train_res),axis=1)
```

```python
test_df  = df.sample(100000)
```

```python
test_df.shape
```

```python
from pycaret.classification import *
setup(test_df,target='Class')
```

```python
compare_models()
```

```python
test_df['Class'].value_counts() 
```

```python
create_model('et')
```

## Model Training and Testing with all data 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_res, y_train_res, test_size=0.30, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

classifier = ExtraTreesClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.predict(X_test)
```

```python
prediction =classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score

print('******confusion_matrix******* ')
print(confusion_matrix(y_test,prediction))
print('******accuracy_score******* ')
print(accuracy_score(y_test,prediction))
print('******classification_report******* ')
print(classification_report(y_test,prediction))
print('******auc_score******* ')
print(roc_auc_score(y_test,prediction))
```
